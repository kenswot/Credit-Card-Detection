```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
birth = pd.read_csv('Births2015.csv', index_col=[0], parse_dates = [0], squeeze = True)
```

```python
birth_mean.plot()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
sns.countplot(birth['births'].iloc[1:10] , palette = 'pink')
plt.title('Birth From 1 to 10', fontsize = 20)
plt.show()
```

```python
birth['month'].value_counts().plot.bar(color = 'orange', figsize = (20, 7));
plt.title('Month in Bar Chart', fontsize = 30, fontweight = 20)
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()
```

```python
# Plot histograms of each parameter 
birth.hist(figsize = (20, 20))
plt.show()
```
